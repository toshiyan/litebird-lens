```python
import numpy as np, prjlib, plottools as pl
from matplotlib.pyplot import *
```

```python
p = prjlib.init_analysis(stype='nkap12')
g = prjlib.init_gal(p.ids,p.stype)
l, gg, gp, pp = np.loadtxt(g.mcl,unpack=True,usecols=(0,1,2,3))
```

```python
pl.plot_1dstyle(fsize=[8,6],xmin=2,xmax=4096,xlog=True,ylog=True)
plot(l,gg,label='gg')
plot(l,l**2*gp,label='gp')
plot(l,l**4*pp,label='pp')
legend()
```
