```python
import numpy as np, local, healpy as hp
from matplotlib.pyplot import *
import warnings
warnings.filterwarnings("ignore")
```

```python
glob  = local.analysis()
W = {}
W['litebird'] = hp.read_map(glob.wind['litebird'])
for survey in ['euclid','lsst','cib','cmbs4']:
    W[survey] = hp.read_map(glob.wind[survey])
```

```python
for wname, title in zip(['litebird','euclid','lsst','cmbs4'],['LiteBIRD / CIB','Euclid','LSST','CMB-S4']):
    matplotlib.rcParams.update({'font.size':17})
    hp.mollview(W[wname],title=title)
    savefig('fig_'+wname+'.pdf')
```
